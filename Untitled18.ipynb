```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the traffic accident dataset
accidents_df = pd.read_csv("C:\\Users\\HARITHA\\Documents\\RTA Dataset.csv")

# Perform data preprocessing if necessary
# For example, convert datetime columns to datetime objects

# Exploratory Data Analysis (EDA)
# Summary statistics
print(accidents_df.describe())

# Visualize accident counts by road conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Road_surface_conditions', data=accidents_df)
plt.title('Accident Counts by Road Conditions')
plt.xlabel('Road Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# Visualize accident counts by weather conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather_conditions', data=accidents_df)
plt.title('Accident Counts by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# Visualize accident counts by time of day
accidents_df['Time'] = pd.to_datetime(accidents_df['Time']).dt.hour
plt.figure(figsize=(10, 6))
sns.histplot(data=accidents_df, x='Time', bins=24, kde=False)

plt.title('Accident Counts by Time of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.xticks(range(0, 24))
plt.show()

plt.figure(figsize=(12, 8))
accidents_df['Area_accident_occured'].value_counts().head(20).plot(kind='barh')
plt.title('Top 20 Accident Hotspots by Street')
plt.xlabel('Number of Accidents')
plt.ylabel('Street Name')
plt.gca().invert_yaxis()  # Invert y-axis to display top accident hotspots at the top
plt.show()
```
